```
#importing the libraries
import torch
import numpy as np
import matplotlib.pyplot as plt
```

```
#dataset
x = np.arange(1,361,1)
y = np.sin(np.radians(x))
print(f'shape of x is {x.shape}')
print(f'shape of y is {y.shape}')
```

```
#sine wave
plt.plot(x,y)
```

```
#converting plain data to machine understandable format
x_train = []
y_train = []

for i in x:
  row = []
  for j in range(i,i+5):
    row.append(j)
  y_train.append(np.sin(np.radians(j)))
  x_train.append(row)
```

```
#converting to numpy arrays and then to tensors
x_train = np.array(x_train)
x_train = torch.tensor(x_train)
y_train = np.array(y_train)
y_train = torch.tensor(y_train)
print(f'shape of x is {x_train.shape} and shape of y is {y_train.shape}')
```
